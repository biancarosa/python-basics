# Trabalhando com arquivos

```python
filename = '100 Sales Records.csv'
```

```python
with open(filename) as file:
    print(file)
```

```python
with open(filename) as file:
    print(file.read())
```

```python
with open(filename) as file:
    for line in file:
        print(line)
```

```python
with open(filename) as file:
    for line in file:
        print(line.strip())
```

```python
lista = []
with open(filename) as file:
    for line in file:
        splitted_line = line.strip().split(',')
        lista.append({
            'Region' : splitted_line[0], 
            'Country' : splitted_line[1], 
            'Item Type' : splitted_line[2], 
            'Sales Channel' : splitted_line[3], 
            'Order Priority' : splitted_line[4], 
            'Order Date' : splitted_line[5], 
            'Order ID' : splitted_line[6], 
            'Ship Date' : splitted_line[7], 
            'Units Sold' : splitted_line[8], 
            'Unit Price' : splitted_line[9], 
            'Unit Cost' : splitted_line[10], 
            'Total Revenue' : splitted_line[11], 
            'Total Cost' : splitted_line[12], 
            'Total Profit' : splitted_line[13]
        })
print(lista)
```
